```python
import matplotlib.pyplot as plt

models = [
    'DT-3',
    'DT-5',
    'DT-10',
    'Random Forest',
    'Logistic Regression',
    'KNN',
    'Naive Bayes',
    'SVM'
]

accuracies = [
    74.28,
    86.99,
    95.38,
    97.40,
    65.90,
    88.73,
    62.43,
    91.33
]

plt.figure(figsize=(10,5))
plt.bar(models, accuracies)

plt.title("Model Accuracy Comparison")
plt.ylabel("Accuracy (%)")
plt.xticks(rotation=45)

plt.show()
```
