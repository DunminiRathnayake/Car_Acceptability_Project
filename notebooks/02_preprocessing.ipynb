```python
import pandas as pd

columns = [
    'buying',
    'maint',
    'doors',
    'persons',
    'lug_boot',
    'safety',
    'class'
]

df = pd.read_csv('../data/car.csv', names=columns)

df.head()
```

```python
for col in df.columns:
    print(col)
    print(df[col].unique())
    print()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in df.columns:
    df[col] = le.fit_transform(df[col])

df.head()
```

```python
df.info()
```

```python
df.to_csv('../data/car_encoded.csv', index=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

sns.heatmap(df.corr(), annot=True)

plt.title("Correlation Heatmap")
plt.show()
```
