```python
import pandas as pd

df = pd.read_csv('../data/car_encoded.csv')

print(df.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,5))

sns.countplot(x='class', data=df)

plt.title('Car Acceptability Classes')

plt.show()
```

Observation:

Most cars belong to the unacceptable category. The dataset is imbalanced.

```python
plt.figure(figsize=(8,5))

sns.countplot(x='safety', hue='class', data=df)

plt.title('Safety vs Car Acceptability')

plt.show()
```

Observation:

Cars with better safety ratings tend to have better acceptability classifications.

```python
plt.figure(figsize=(8,5))

sns.countplot(x='buying', hue='class', data=df)

plt.title('Buying Price vs Car Acceptability')

plt.show()
```

Observation:

Buying price influences car acceptability. Different price categories show different class distributions.

```python
plt.figure(figsize=(8,5))

sns.countplot(x='maint', hue='class', data=df)

plt.title('Maintenance Cost vs Car Acceptability')

plt.show()
```

Observation:

Maintenance cost affects car acceptability. Cars with lower maintenance costs tend to be more acceptable.

```python
plt.figure(figsize=(8,6))

sns.heatmap(df.corr(), annot=True)

plt.title('Correlation Heatmap')

plt.show()
```

Observation:

The heatmap shows relationships between features and the target variable.
